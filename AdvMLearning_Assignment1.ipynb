```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.datasets import imdb
from keras import models, layers

# Load the IMDB dataset
(training_data, training_labels), (testing_data, testing_labels) = imdb.load_data(num_words=10000)

# Vectorize the data
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

x_training = vectorize_sequences(training_data)
x_testing = vectorize_sequences(testing_data)

# Vectorize labels
y_training = np.asarray(training_labels).astype('float32')
y_testing = np.asarray(testing_labels).astype('float32')

# Validation set
x_validation = x_training[:10000]
partial_x_training = x_training[10000:]
y_validation = y_training[:10000]
partial_y_training = y_training[10000:]

# Experiment 1: Single Hidden Layer
```

```python
model1 = models.Sequential()
model1.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model1.add(layers.Dense(1, activation='sigmoid'))
model1.compile(optimizer='rmsprop',
               loss='binary_crossentropy',
               metrics=['accuracy'])
history1 = model1.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results1 = model1.evaluate(x_testing, y_testing)
print("Experiment 1 Results (Single Hidden Layer):")
print("Test Accuracy:", results1[1])
```

```python
# Experiment 2: Three Hidden Layers
```

```python
model2 = models.Sequential()
model2.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model2.add(layers.Dense(16, activation='relu'))
model2.add(layers.Dense(16, activation='relu'))
model2.add(layers.Dense(1, activation='sigmoid'))
model2.compile(optimizer='rmsprop',
               loss='binary_crossentropy',
               metrics=['accuracy'])
history2 = model2.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results2 = model2.evaluate(x_testing, y_testing)
print("\nExperiment 2 Results (Three Hidden Layers):")
print("Test Accuracy:", results2[1])
```

```python
# Experiment 3: Hidden Layers with 32 Units
model3 = models.Sequential()
model3.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model3.add(layers.Dense(32, activation='relu'))
model3.add(layers.Dense(1, activation='sigmoid'))
model3.compile(optimizer='rmsprop',
               loss='binary_crossentropy',
               metrics=['accuracy'])
history3 = model3.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results3 = model3.evaluate(x_testing, y_testing)
print("\nExperiment 3 Results (Hidden Layers with 32 Units):")
print("Test Accuracy:", results3[1])
```

```python
# Experiment 4: MSE Loss Function
model4 = models.Sequential()
model4.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model4.add(layers.Dense(16, activation='relu'))
model4.add(layers.Dense(1, activation='sigmoid'))
model4.compile(optimizer='rmsprop',
               loss='mse',
               metrics=['accuracy'])
history4 = model4.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results4 = model4.evaluate(x_testing, y_testing)
print("\nExperiment 4 Results (MSE Loss Function):")
print("Test Accuracy:", results4[1])
```

```python
# Experiment 5: Tanh Activation
model5 = models.Sequential()
model5.add(layers.Dense(16, activation='tanh', input_shape=(10000,)))
model5.add(layers.Dense(16, activation='tanh'))
model5.add(layers.Dense(1, activation='sigmoid'))
model5.compile(optimizer='rmsprop',
               loss='binary_crossentropy',
               metrics=['accuracy'])
history5 = model5.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results5 = model5.evaluate(x_testing, y_testing)
print("\nExperiment 5 Results (Tanh Activation):")
print("Test Accuracy:", results5[1])
```

```python
# Experiment 6: Dropout Technique
model6 = models.Sequential()
model6.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model6.add(layers.Dropout(0.5))
model6.add(layers.Dense(16, activation='relu'))
model6.add(layers.Dropout(0.5))
model6.add(layers.Dense(1, activation='sigmoid'))
model6.compile(optimizer='rmsprop',
               loss='binary_crossentropy',
               metrics=['accuracy'])
history6 = model6.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results6 = model6.evaluate(x_testing, y_testing)
print("\nExperiment 6 Results (Dropout Technique):")
print("Test Accuracy:", results6[1])
```

```python
# Experiment 7: Regularization Technique
from keras import regularizers
model7 = models.Sequential()
model7.add(layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), activation='relu', input_shape=(10000,)))
model7.add(layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), activation='relu'))
model7.add(layers.Dense(1, activation='sigmoid'))
model7.compile(optimizer='rmsprop',
               loss='binary_crossentropy',
               metrics=['accuracy'])
history7 = model7.fit(partial_x_training, partial_y_training,
                      epochs=20, batch_size=512,
                      validation_data=(x_validation, y_validation))
results7 = model7.evaluate(x_testing, y_testing)
print("\nExperiment 7 Results (Regularization Technique):")
print("Test Accuracy:", results7[1])
```
